```python
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt
```

```python
fillcolor=(255,255,255)
imgtr=cv2.imread("top_right.jpg")
imgtl=cv2.imread("top_left.jpg")

imgtr=cv2.cvtColor(imgtr,cv2.COLOR_BGR2RGB)
imgtl=cv2.cvtColor(imgtl,cv2.COLOR_BGR2RGB)
```

```python
imgtr=cv2.resize(imgtr,(200,200))
imgtl=cv2.resize(imgtl,(200,200))
imgtr = cv2.copyMakeBorder(imgtr, 10, 10, 0, 10,cv2.BORDER_CONSTANT,value=fillcolor)
imgtl = cv2.copyMakeBorder(imgtl, 10, 10, 0, 10,cv2.BORDER_CONSTANT,value=fillcolor)
#https://docs.opencv.org/3.1.0/d3/df2/tutorial_py_basic_ops.html
```

```python
imgt=cv2.hconcat([imgtl,imgtr])
```

```python
plt.imshow(imgt)
```

```python
imgbr=cv2.imread("bottom_right.jpg")
imgbl=cv2.imread("bottom_left.jpg")

imgbr=cv2.cvtColor(imgbr,cv2.COLOR_BGR2RGB)
imgbl=cv2.cvtColor(imgbl,cv2.COLOR_BGR2RGB)
```

```python
imgbr=cv2.resize(imgbr,(200,200))
imgbl=cv2.resize(imgbl,(200,200))
#top, bottom, left, right 
imgbr=cv2.copyMakeBorder(imgbr, 0, 10, 10, 10,cv2.BORDER_CONSTANT,value=fillcolor)
imgbl=cv2.copyMakeBorder(imgbl, 0, 10, 10, 10,cv2.BORDER_CONSTANT,value=fillcolor)
```

```python
imgb=cv2.hconcat([imgbl,imgbr])
```

```python
plt.imshow(imgb)
```
