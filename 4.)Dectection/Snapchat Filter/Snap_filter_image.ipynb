```python
img = cv2.imread("./Test/Before.png")
eyes_cascade=cv2.CascadeClassifier("Train/third-party/frontalEyes35x16.xml")
nose_cascade=cv2.CascadeClassifier("Train/third-party/Nose18x15.xml") 

eyes=eyes_cascade.detectMultiScale(img, 1.3, 5)
nose=nose_cascade.detectMultiScale(img,1.5,5)



xe,ye,we,he=eyes[0]
specs=cv2.imread("./Train/glasses.png",-1)
specs=cv2.resize(specs,(we,he))

for i in range(specs.shape[0]):
    for j in range(specs.shape[1]):
        if (specs[i,j,3]>0):
            img[ye+i,xe+j,:]=specs[i,j,:-1]
            

           
        
xn,yn,wn,hn=nose[0]
musk = cv2.imread("./Train/mustache.png",-1)
musk=cv2.resize(musk,(wn,hn))
            
for i in range(musk.shape[0]):
    for j in range(musk.shape[1]):
        if(musk[i,j,3]>0):
            img[yn+28+i,xn+5+j,:]=musk[i,j,:-1]
            
            
            
         
        
        
musk_on_face=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(musk_on_face)
plt.show()




```
