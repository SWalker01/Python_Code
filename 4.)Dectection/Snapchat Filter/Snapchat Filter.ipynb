```python
import numpy as np
import os
import cv2

frame1=cv2.imread("Jamie_Before.jpg")

specs=cv2.imread('glasses.png')
mustaches=cv2.imread('mustache.png')

face_cascade=cv2.CascadeClassifier("haarcascade_frontalface_alt.xml")
eyes_cascade=cv2.CascadeClassifier("frontalEyes35x16.xml")


while True:
    #ret,frame=cap.read()
    frame = cv2.cvtColor(frame1, cv2.COLOR_BGR2GRAY)
    
    #if ret==False:
       # continue
        
    faces=face_cascade.detectMultiScale(frame,1.3,5)
    eyes=eyes_cascade.detectMultiScale(frame,1.3,5)
    noses=nose_cascade.detectMultiScale(frame,1.3,5)
    for eye in eyes:
        (ex,ey,ew,eh)=eye
        specs=cv2.resize(specs,(ew,eh))
        sw,sh,sc=specs.shape
        for i in range(0,sw):
            for j in range(0,sh):
                if specs[i][j][3]!=0:
                    frame[ey+i][ex+j]=specs[i][j]
    for nose in noses:
        (nx,ny,nw,nh)=nose
        mustaches=cv2.resize(mustaches,(nw,nh))
        mw,mh,mx=mustaches.shape
        for i in range(0,mw):
            for j in range(0,mh):
                if mustaches[i][j][3]!=0:
                    frame[ny+i][nx+j]=mustaches[i][j]

    cv2.imshow("Frame",frame)
    key_pressed=cv2.waitKey(1) & 0xFF
    if key_pressed==ord('q'):
        break

        
cap.release()
cv2.destroyAllWindows()
```
