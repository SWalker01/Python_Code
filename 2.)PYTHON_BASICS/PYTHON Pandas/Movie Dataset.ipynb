```python
freq_title={}

for t in title:
    length=len(t)
    
    try:#u can use if else also
        freq_title[length]+=1
    except:
        freq_title[length]=1
        
print(freq_title)        
```

```python
#plot for the above Dataset
X=np.array(list(freq_title.keys()))
y=np.array(list(freq_title.values()))
```

```python
plt.scatter(X,y)
plt.xlabel("length of all Movie Title in Dataset")
plt.ylabel("no. of movies of the length")
```
