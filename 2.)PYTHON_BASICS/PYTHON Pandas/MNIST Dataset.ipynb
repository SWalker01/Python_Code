```python
X=data[:,1:]
y=data[:,0]
```

```python
print(X.shape,y.shape)
```

```python
#to vizualize the img we have to convert flatten img into 2D img
```

```python
def drwimg(X,y,i):
    plt.imshow(X[i].reshape(28,28))
    plt.show()
    
    
drwimg(X,y,0)    
```

```python
def drwimg(X,y,i):
    plt.imshow(X[i].reshape(28,28),cmap="gray")
    plt.title("Label"+str(y[i])) #which no. img represent
    plt.show()
    
    
drwimg(X,y,0)    
```

```python
#you can also use sklearn libarary
from sklearn.model_selection import train_test_split

XT,Xt,yT,yt= train_test_split(X,y,test_size=0.2,random_state=5)
```

```python
print(XT.shape,yT.shape)
print(Xt.shape,yt.shape)
```

```python
for i in range(5):
    drwimg(X,y,i)    
    
```
