# Normalise

```python
dfx =pd.read_csv("../../Datasets/linearX.csv").values
dfy= pd.read_csv("../../Datasets/linearY.csv").values

x=dfx
y=dfy

#normalize

x2=(x-x.mean())/x.std()





plt.scatter(x,y,label="Raw Data")
plt.scatter(x2,y,c="orange",label="Normalized Data")
plt.legend()
plt.show()
```
